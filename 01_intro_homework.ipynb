```python
selected_columns = loc_island.loc[:, ['housing_median_age', 'total_rooms', 'total_bedrooms']]
selected_columns
```

```python
X = np.array(selected_columns)
X
```

```python
XTX = X.T.dot(X)
XTX
```

```python
XTX_inverse = np.linalg.inv(XTX)
XTX_inverse
```

```python
y = np.array([950, 1300, 800, 1000, 1300])
```

```python
XTX_inverse_XT = XTX_inverse.dot(X.T)
XTX_inverse_XT
```

```python
w = XTX_inverse_XT.dot(y)
w
```
